```python
from datasets import Dataset
import os

def convert_test_file_to_dataset(file_path):
    problem_list = []

    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line:
                # Thêm dấu chấm nếu chưa có
                if not line.endswith("."):
                    line += "."
                problem_list.append(line)

    return Dataset.from_dict({
        "problem": problem_list
    })
```

```python
import asyncio
import re
from tqdm.asyncio import tqdm_asyncio  # Dùng để hiển thị tqdm với async
from typing import List
import nest_asyncio

nest_asyncio.apply()  # Cho phép chạy event loop lồng nhau (Colab/Jupyter)

async def generate_one(sample_problem: str, tokenizer, model) -> str:
    messages = [
        {"role": "system", "content": "You are a helpful assistant,"},
        {"role": "user", "content": sample_problem}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=True,
    )

    inputs = tokenizer(text, return_tensors="pt").to("cuda")
    input_length = inputs['input_ids'].shape[1]

    output_ids = model.generate(
        **inputs,
        max_new_tokens=4000,
        temperature=0.2,
        top_p=0.95,
        top_k=20,
    )
    gen_tokens = output_ids[0][input_length:]

    assistant_text = tokenizer.decode(gen_tokens, skip_special_tokens=True).strip()
    assistant_text = re.sub(r"<think>.*?</think>\s*", "", assistant_text, flags=re.DOTALL)
    assistant_text = assistant_text.replace("<|im_end|>", "").strip()
    return assistant_text

async def process_all(valid_dataset, tokenizer, model, output_file: str, batch_size: int = 8):
    tasks = []
    for item in valid_dataset:
        sample_problem = item["problem"]
        tasks.append(generate_one(sample_problem, tokenizer, model))

    results = []
    for i in tqdm_asyncio(range(0, len(tasks), batch_size), desc="Processing"):
        batch = tasks[i:i + batch_size]
        batch_results = await asyncio.gather(*batch)
        results.extend(batch_results)

    with open(output_file, "w", encoding="utf-8") as f:
        for i, result in enumerate(results):
            sentence = valid_dataset[i]["problem"]
            f.write(f"#::snt {sentence.strip()}\n")
            f.write(f"{result.strip()}\n\n")

await process_all(valid_dataset, tokenizer, model, "output.txt")
```
