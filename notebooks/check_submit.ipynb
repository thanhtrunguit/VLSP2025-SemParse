```python
import os
from amr import AMR

def multiline_to_oneline(text: str) -> str:
    return " ".join(line.strip() for line in text.strip().splitlines())

def parse_amr_no_check(amr_multiline: str):
    try:
        amr_line = multiline_to_oneline(amr_multiline)
        return AMR.parse_AMR_line(amr_line)
    except Exception:
        return None

def process_amr_files(folder_path: str):
    count_error = 0  # Bi·∫øn ƒë·∫øm l·ªói

    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, "r", encoding="utf-8") as f:
                content = f.read()

            samples = content.split("#::snt ")
            for sample in samples[1:]:  # sample[0] is likely empty
                try:
                    sentence, *amr_lines = sample.strip().splitlines()
                    amr_multiline = "\n".join(amr_lines)
                    amr = parse_amr_no_check(amr_multiline)
                    if amr is None:
                        count_error += 1
                        print("‚ùå Kh√¥ng parse ƒë∆∞·ª£c:")
                        print(">>", sentence)
                except Exception as e:
                    count_error += 1
                    print("‚ùå‚ùå L·ªói ƒë·ªãnh d·∫°ng ho·∫∑c parse:", e)
                    print(">>", sentence[:100])  # ch·ªâ in 100 k√Ω t·ª± ƒë·∫ßu n·∫øu l·ªói

    print(f"\nüëâ T·ªïng s·ªë l·ªói AMR parse: {count_error}")
```

```python
#bo _
import re

def is_proper_name(word):
    parts = word.split("_")
    return all(part and part[0].isupper() for part in parts)

def replace_underscores(text):
    def replace_token(match):
        token = match.group()
        if is_proper_name(token):
            return token  # gi·ªØ nguy√™n
        else:
            return token.replace("_", " ")  # thay _ b·∫±ng kho·∫£ng tr·∫Øng

    # Match c√°c t·ª´ c√≥ ch·ª©a d·∫•u _
    return re.sub(r'\w+_\w+(?:_\w+)*', replace_token, text)

def process_amr_file(input_path, output_path):
    with open(input_path, 'r', encoding='utf-8') as f:
        content = f.read()

    # T√°ch th√†nh c√°c block: m·ªói block g·ªìm #::snt + c√¢u + graph
    blocks = content.strip().split("#::snt ")
    processed = []

    for block in blocks:
        if not block.strip():
            continue
        try:
            snt, *graph_lines = block.strip().split("\n")
            graph = "\n".join(graph_lines)
            graph_cleaned = replace_underscores(graph)
            processed.append(f"#::snt {snt}\n{graph_cleaned}")
        except Exception as e:
            print("‚ö†Ô∏è L·ªói khi x·ª≠ l√Ω block:", block[:50])
            print(e)
            continue

    with open(output_path, "w", encoding="utf-8") as f:
        f.write("\n\n".join(processed))

# G·ªçi h√†m x·ª≠ l√Ω
process_amr_file("/content/results.txt", "output.txt")
```
